```python
def get_comments(video_id):
    comments = []
    nextPageToken = None
    processed_comments = set()  # Sử dụng một set để lưu trữ các comment đã được xử lý
    while True:
        comment_threads = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            textFormat='plainText',
            pageToken=nextPageToken
        ).execute()

        for comment in comment_threads['items']:
            comment_id = comment['id']
            if comment_id in processed_comments:
                continue  # Bỏ qua comment đã được xử lý trước đó

            snippet = comment['snippet']['topLevelComment']['snippet']
            author = snippet['authorDisplayName']
            author_channel_url = snippet['authorChannelUrl']
            content = snippet['textDisplay']
            published_at = snippet['publishedAt']

            comment_info = {
                'author': author,
                'authorChannelUrl': author_channel_url,
                'content': content,
                'published_at': published_at
            }

            comments.append(comment_info)
            processed_comments.add(comment_id)  # Đánh dấu comment đã được xử lý

            # Lấy replies của comment
            reply_comment_id = comment_id
            replies = youtube.comments().list(
                part='snippet',
                parentId=reply_comment_id,
                maxResults=100
            ).execute()

            for reply in replies['items']:
                reply_id = reply['id']
                if reply_id in processed_comments:
                    continue  # Bỏ qua reply đã được xử lý trước đó

                reply_snippet = reply['snippet']
                reply_author = reply_snippet['authorDisplayName']
                reply_author_channel_url = reply_snippet['authorChannelUrl']
                reply_content = reply_snippet['textDisplay']
                reply_published_at = reply_snippet['publishedAt']

                reply_info = {
                    'author': reply_author,
                    'authorChannelUrl': reply_author_channel_url,
                    'content': reply_content,
                    'published_at': reply_published_at
                }

                comments.append(reply_info)
                processed_comments.add(reply_id)  # Đánh dấu reply đã được xử lý

        nextPageToken = comment_threads.get('nextPageToken')
        if not nextPageToken:
            break

    return comments


video_comments = get_comments(video_id)

```

```python
import requests

class YouTubeCommentFetcher:
    def __init__(self, api_key):
        self.api_key = api_key

    def fetch_comments(self, video_id):
        comments = []
        nextPageToken = None
        processed_comments = set()

        while True:
            base_url = 'https://www.googleapis.com/youtube/v3/commentThreads'
            params = self._build_params(video_id, nextPageToken)

            response = requests.get(base_url, params=params)
            data = response.json()

            for comment in data['items']:
                comment_id = comment['id']
                if comment_id in processed_comments:
                    continue

                comment_info = self._process_comment(comment)
                comments.append(comment_info)
                processed_comments.add(comment_id)

                reply_comment_id = comment_id
                params['parentId'] = reply_comment_id
                response = requests.get(base_url, params=params)
                reply_data = response.json()

                self._process_replies(reply_data, comments, processed_comments)

            nextPageToken = data.get('nextPageToken')
            if not nextPageToken:
                break

        return comments

    def _build_params(self, video_id, page_token):
        return {
            'part': 'snippet',
            'videoId': video_id,
            'textFormat': 'plainText',
            'key': self.api_key,
            'maxResults': 100,
            'pageToken': page_token
        }

    def _process_comment(self, comment):
        snippet = comment['snippet']['topLevelComment']['snippet']
        author = snippet['authorDisplayName']
        author_channel_url = snippet['authorChannelUrl']
        content = snippet['textDisplay']
        published_at = snippet['publishedAt']

        return {
            'author': author,
            'authorChannelUrl': author_channel_url,
            'content': content,
            'published_at': published_at
        }

    def _process_replies(self, reply_data, comments, processed_comments):
        for reply in reply_data['items']:
            reply_id = reply['id']
            if reply_id in processed_comments:
                continue

            reply_info = self._process_comment(reply)
            comments.append(reply_info)
            processed_comments.add(reply_id)

# Sử dụng lớp để lấy danh sách comment
api_key = 'AIzaSyBZnBVHK5L57k4cwbEVPN7GXMKHOlmnW3U'
video_id = '-FITFOdvQLw'
comment_fetcher = YouTubeCommentFetcher(api_key)
comments = comment_fetcher.fetch_comments(video_id)

for comment in comments:
    print(comment)
```

```python
import requests
import json

class YouTubeAPI:
    def __init__(self, api_key, channel_id):
        self.api_key = api_key
        self.channel_id = channel_id
        self.base_url = 'https://www.googleapis.com/youtube/v3/'

    def get_channel_videos(self, max_results=5):
        videos = self._get_channel_videos(max_results)
        video_comments = []

        for video in videos:
            video_id = video['id']['videoId']
            comments = self._get_video_comments(video_id)
            
            video_comments.append({
                'video_id': video_id,
                'comments': [comment['snippet']['topLevelComment']['snippet']['textDisplay'] for comment in comments]
            })

        return video_comments

    def _get_channel_videos(self, max_results, page_token=None):
        params = {
            'key': self.api_key,
            'channelId': self.channel_id,
            'part': 'id',
            'order': 'date',
            'maxResults': max_results,
            'pageToken': page_token
        }

        response = requests.get(self.base_url + 'search', params=params)
        result = response.json()
        videos = result.get('items', [])

        # Check if there is a next page and retrieve it
        next_page_token = result.get('nextPageToken')
        if next_page_token:
            more_videos = self._get_channel_videos(max_results, next_page_token)
            videos.extend(more_videos)

        return videos

    def _get_video_comments(self, video_id, max_results=10):
        comments_params = {
            'key': self.api_key,
            'part': 'snippet',
            'videoId': video_id,
            'order': 'time',
            'textFormat': 'plainText',
            'maxResults': max_results
        }

        response = requests.get(self.base_url + 'commentThreads', params=comments_params)
        comments = response.json().get('items', [])
        return comments

# Thay thế 'YOUR_API_KEY' và 'YOUR_CHANNEL_ID' bằng thông tin của bạn
channel_id = 'UCbXgNpp0jedKWcQiULLbDTA'
api_key = 'AIzaSyB6Qhlr3A0YReR5f_imkwgv7L-xDlVPDQw'

# Tạo một đối tượng YouTubeAPI
youtube_api = YouTubeAPI(api_key, channel_id)

# Lấy và in kết quả dưới dạng JSON
result = youtube_api.get_channel_videos()

```
